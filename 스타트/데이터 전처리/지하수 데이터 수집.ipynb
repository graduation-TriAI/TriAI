```python
import requests
import json
from urllib.parse import unquote

# usl
url = "http://www.gims.go.kr/api/data/observationStationService/getGroundwaterMonitoringNetwork"

# 2. 키 디코딩
encoded_key = KEY
decoded_key = unquote(encoded_key)

# 3. params
params = {
    "KEY": decoded_key,
    "type": "json",
    "gennum": "622083",
    "begindate": "20170101",
    "enddate": "20171231"
}

# 4. 요청
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    print(json.dumps(response.json(), indent=4, ensure_ascii=False))
    file_name = f"/content/drive/MyDrive/TriAI/Data/포항_201711/지하수 데이터/gims_data_622083.json"

    with open(file_name, 'w', encoding='utf-8') as f:
      json.dump(data, f, indent=4, ensure_ascii=False)
else:
    print("Error:", response.status_code)
```
